# Best starbucks options for people looking to get jacked
* Audience: non-technical consumers or marketers

There are many people in the world looking to get into great shape. To that end they will need food options that support this goal. When going to starbucks, they would most likely want to find drinks with high caffein to fuel their workouts. They will also want to find food and drinks with high protein to replenish their muscles. People who care only about adding mass, called "dirty bulkers" will seek high protein food and will not care about extra calories. Those who want to add lean body mass, "clean bulkers", will seek out food where protein has a high ratio to the total calories in the food.

```python
drinks = pd.read_csv('starbucks-menu-nutrition-drinks.csv')
food = pd.read_csv('starbucks-menu-nutrition-food.csv', encoding='utf-16')
drinks_exp = pd.read_csv('starbucks_drinkMenu_expanded.csv')
```

* Interestign columns will be those concerning calories, caffeine, protein, carbs, fat


```python
food.rename(columns = {'Unnamed: 0': 'item'}, inplace=True)
drinks.rename(columns = {'Unnamed: 0': 'item'}, inplace=True)
display(drinks.head(3), food.head(3), drinks_exp.head())
```

```python
display(drinks.info(3), food.info(3), drinks_exp.info())
```

```python
drink = drinks_exp
```

### Data Cleaning
Here we clean up the column names and take out rows where caffeine is not a number

```python
pythonic_list = [
    'beverage_category', 'beverage', 'beverage_prep', 'calories',
    'total_fat', 'trans_fat', 'saturated_fat', 'sodium',
    'total_carbohydrates', 'cholesterol', 'dietary_fibre',
    'sugars', 'protein', 'vitamin_a_%_dv', 'vitamin_c_%_dv',
    'calcium_%_dv', 'iron_%_dv', 'caffeine_mg'
]
pythonic_list_food = ['item', 'calories', 'fat_g', 'carb_g', 'fiber_g', 'protein_g']
drink.columns = pythonic_list
food.columns = pythonic_list_food
```

```python
mask = (drink.caffeine_mg != 'Varies') & (drink.caffeine_mg != 'varies') & (drink.caffeine_mg.notnull())
drink = drink[mask]
drink.caffeine_mg = drink.caffeine_mg.astype('int')
```

## Are there any high protein drinks? 

```python
drink.protein.hist(bins=20)
plt.title("Protein content distribution in all starbucks drinks")
plt.show()
```

```python
p = drink[drink.protein >= 15].drop_duplicates('beverage').sort_values('protein', ascending=False)
sns.barplot(p, x='protein', y='beverage')
plt.title('Top drinks for high protein content')
plt.xlabel('protein in grams')
plt.xticks(rotation=-90);
```

## What is the best choice for people who want the most caffeine? 

```python
drink.caffeine_mg.hist(bins=20)
plt.title("caffeine_mg content distribution in all starbucks drinks")
plt.xticks(rotation=-90);
```

```python
c = drink[drink.caffeine_mg >= 200].drop_duplicates('beverage').sort_values('caffeine_mg', ascending=False)
sns.barplot(c, x='caffeine_mg', y='beverage')
plt.title('Top drinks for high caffeine content')
plt.xlabel('caffeine in milligrams')
plt.xticks(rotation=-90);
```

## What food items have the highest protein content

```python
pro = food.sort_values('protein_g', ascending=False)
sns.barplot(pro.iloc[:10], x='protein_g', y='item')
plt.title('Top high protein for dirty bulkers')
plt.xlabel('protein in grams')
plt.xticks(rotation=-90);
```

## What food items have the best ratio of calories from protein to total calories?

```python
for col, cal_per_g in zip(['fat_g', 'carb_g', 'protein_g'], [9, 4, 4]): 
    food[col[0]] = (cal_per_g * food[col] * 100).astype(int)
food['p_ratio'] = food['p'] / food.calories
pro = food.sort_values('p_ratio', ascending=False)
sns.barplot(pro.iloc[:10], x='p_ratio', y='item')
plt.title('Top foods for high protein/total_calorie ratio')
plt.xlabel('calories from protein percentage')
plt.xticks(rotation=-90);
```

## Conclusion
* Best for caffeine:  brewed coffee, iced brewed coffee, or a caffe americano
* Best for Dirty bulkers: 'Turkey Pesto Panini', 'Roasted Turkey & Dill Havarti Sandwich' or a 'Turkey & Havarti Sandwich'. 
* Best for Clean bulkers: 'Sous Vide Egg Bites: Egg White & Red Pepper', 'Reduced-Fat Turkey Bacon & Cage Free Egg White Breakfast Sandwich' or 'Smoked Turkey Protein Box'.
